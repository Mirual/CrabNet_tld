**Get data from materials project**

```python
import keys
from mp_api.client import MPRester

with MPRester(keys.mp_api) as mpr:
    ## -- Alternative directly using energy above hull:
    docs = mpr.materials.summary.search(
        band_gap=(6, None), energy_above_hull=(0, 0), fields=["material_id", "band_gap", "formula_pretty"]
    
    )
    stable_mpids = [[doc.material_id, doc.formula_pretty, doc.band_gap] for doc in docs]
    print(stable_mpids)
```

**Train, val, test split to files**

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = stable_mpids
X_train, X_val_test = train_test_split(X, test_size=0.25, random_state=42)
X_val, X_test = train_test_split(X_val_test, test_size=0.10, random_state=42)
print(X_train)
print(X_val)
print(X_test)

X_train_csv = [x[1:] for x in X_train]
X_val_csv = [x[1:] for x in X_val]
X_test_csv = [x[1:] for x in X_test]
```

**Save data to csv for training**

```python
import csv
fields = ["formula", "target"]
# TRAIN
with open('data/materials_data/train.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
     
    write.writerow(fields)
    write.writerows(X_train_csv)
 
# VAL
with open('data/materials_data/val.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
     
    write.writerow(fields)
    write.writerows(X_val_csv)   
    
# TEST
with open('data/materials_data/test.csv', 'w') as f:
     
    # using csv.writer method from CSV package
    write = csv.writer(f)
     
    write.writerow(fields)
    write.writerows(X_test_csv)
```
